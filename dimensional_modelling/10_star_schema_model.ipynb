# Designing the star schema

## Grain
`One row in fct_results = one athlete result in one event`

**Fact Table:**
`fact_results`

**Dimension Tables:** `dim_event, dim_athlete, dim_country, dim_date`

## fact_results
```
  result_id string
  event_id string
  athlete_id integer
  country_code_iso3 string
  date_id integer
  event_number_of_finishers integer
  athlete_performance string
  athlete_performance_type string
  athlete_performance_seconds integer
  athlete_performance_distance double
  athlete_performance_unit string
  athlete_average_speed string
  athlete_average_speed_kmh double
  athlete_age_at_event integer
```

## dim_event
```
  event_id string 
  event_name string
  event_distance_length string
  event_distance_type string
  event_distance_value double
  event_distance_unit string
```

## dim_athlete
```
  athlete_id integer
  athlete_gender string
  athlete_year_of_birth integer
  athlete_age_category string
  athlete_club string
```

## dim_country
```
  country_code_iso3 string
  athlete_country string
  country_name string
```

## dim_date
```
  date_id integer
  event_start_date date
  year integer
  month integer
  month_name string
```
