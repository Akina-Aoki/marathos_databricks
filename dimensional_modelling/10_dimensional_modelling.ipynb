# Designing the star schema


# Conceptual Model

```
Event    1 ───< Result
Athlete  1 ───< Result
Country  1 ───< Result
Date     1 ───< Result
```

One Event can have many Results.
Each Result belongs to one Event.

One Athlete can have many Results.
Each Result belongs to one Athlete.

One Country can have many Results.
Each Result is linked to one Country.

One Date can have many Results.
Each Result is linked to one Date.

![Conceptual Model](marathos_conceptual_model.png)

# Logical Model

- Result is the fact-like central entity because it contains the measurable race outcome.

- athlete_age_at_event stays in Result because age depends on the event year.

- athlete_year_of_birth stays in Athlete because birth year is stable athlete information.

- date_id is created from event_start_date in the Gold layer.

- Country is connected to Result because the country value comes from the recorded athlete result and supports country-based analysis.

# Physical Model

![Physical Model](physical_model.png)

## Grain
`One row in fct_results = one athlete result in one event`

**Fact Table:**
`fact_results`

**Dimension Tables:** `dim_event, dim_athlete, dim_country, dim_date`

## fact_results
```
  result_id string
  event_id string
  athlete_id integer
  country_code_iso3 string
  date_id integer
  event_number_of_finishers integer
  athlete_performance string
  athlete_performance_type string
  athlete_performance_seconds integer
  athlete_performance_distance double
  athlete_performance_unit string
  athlete_average_speed string
  athlete_average_speed_kmh double
  athlete_age_at_event integer
```

## dim_event
```
  event_id string 
  event_name string
  event_distance_length string
  event_distance_type string
  event_distance_value double
  event_distance_unit string
```

## dim_athlete
```
  athlete_id integer
  athlete_gender string
  athlete_year_of_birth integer
  athlete_age_category string
  athlete_club string
```

## dim_country
```
  country_code_iso3 string
  athlete_country string
  country_name string
```

## dim_date
```
  date_id integer
  event_start_date date
  year integer
  month integer
  month_name string
```



## About the `dim_date` plan

```
Silver OBT event_start_date
        ↓
Gold dim_date date_id
        ↓
Gold fact_results date_id
```


- `date_id` is a Gold layer / dimensional modelling column that I need to create from the `event_start_date` column created at the silver layer.

- Then in Gold, create dim_date like this:

``` 
event_start_date = 1991-12-07
date_id = 19911207
year = 1991
month = 12
month_name = December
```
